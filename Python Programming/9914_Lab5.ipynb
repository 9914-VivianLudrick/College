```python
def count_lines_words_characters(filename):
  with open(filename, 'r') as f:
    list_of_lines = f.readlines()
    lines = len(list_of_lines)
    f.seek(0) # puts the file pointer to the start
    words = len(f.read().split()) # splits by spaces
    characters = sum(len(line) for line in list_of_lines)

  return lines, words, characters

# Example usage
filename = "new.txt"
lines, words, characters = count_lines_words_characters(filename)

print(f"Lines: {lines}")
print(f"Words: {words}")
print(f"Characters: {characters}")
```

```python
def count_vowels_consonants(filename):
  with open(filename, "r") as f:
    data = f.read()

  # Count the number of vowels and consonants
  vowels = 0
  consonants = 0
  for char in data:
    if char.lower() in "aeiou":
      vowels += 1
    elif char.isalpha():
      consonants += 1

  # Calculate the percent count
  total_chars = len(data)
  percent_vowels = (vowels / total_chars) * 100
  percent_consonants = (consonants / total_chars) * 100

  # Return the results
  return {
    "percent_vowels": percent_vowels,
    "percent_consonants": percent_consonants
  }

# Example usage
filename = "new.txt"
results = count_vowels_consonants(filename)
print("Percent vowels:", round(results["percent_vowels"],3))
print("Percent consonants:", round(results["percent_consonants"],3))
```

```python
def merge_template(template_file, data_file, output_file_base_name):
    with open(template_file, "r") as template:
        template_content = template.read()

    with open(data_file, "r") as data:
        next(data)  # Skip the header line
        for line in data:
            values = line.strip().split(",")
            applicant_name = values[0]
            company_name = values[1]
            position = values[2]
            location = values[3]
            salary = values[4]
            field = values[5]
            interview_type = values[6]
            recruiter = values[7]

            # Replace placeholders with actual values
            merged_content = template_content.replace(
                "{{APPLICANT_NAME}}", applicant_name
            )
            merged_content = merged_content.replace("{{COMPANY_NAME}}", company_name)
            merged_content = merged_content.replace("{{POSITION}}", position)
            merged_content = merged_content.replace("{{LOCATION}}", location)
            merged_content = merged_content.replace("{{SALARY}}", salary)
            merged_content = merged_content.replace("{{FIELD}}", field)
            merged_content = merged_content.replace(
                "{{INTERVIEW_TYPE}}", interview_type
            )
            merged_content = merged_content.replace("{{RECRUITER_NAME}}", recruiter)

            output_file = f"{applicant_name}_{output_file_base_name}"
            # Write merged content to output file
            with open(output_file, "a") as output:
                output.write(merged_content + "\n")

def read_mails(data_file, output_file_base_name):
    with open(data_file, "r") as data:
        next(data) # Skipping the first line
        for line in data:
            values = line.strip().split(",")
            applicant_name = values[0]

            output_file = f"{applicant_name}_{output_file_base_name}"

            print("-"*70)
            print(f"Email received for {applicant_name}: ")
            print("-"*70)
            with open(output_file, "r") as output:
                print(output.read())

# main program
template_file = "template.txt"
data_file = "data.csv"
output_file_base_name = "email.txt"

merge_template(template_file, data_file, output_file_base_name)
print("Emails generated successfully!")
read_mails(data_file, output_file_base_name)
```

```python
def remove_digits(filename):
  """
  This function reads data from a file and writes to another file, excluding digits.

  Args:
    filename: The path to the input file.
  """

  with open(filename, "r") as input_file:
    data = input_file.read()

  # Filter out digits from the data. auto closes the file
  filtered_data = "".join([char for char in data if not char.isdigit()])

  # Write the filtered data to a new file. Auto closes the file
  with open("anotherFile.txt", "w+") as output_file:
    output_file.write(filtered_data)
    output_file.seek(0)
    print(output_file.read())

# Example usage
filename = "new.txt"
remove_digits(filename)
```

```python
def find_longest_word(filename):
  # read file
  with open(filename, 'r') as file:
    # make array of words
    words = file.read().split()

  # check for longest word
  longest_word = ''
  for word in words:
    if len(word) > len(longest_word):
      longest_word = word

  # return the longest word
  return longest_word

# get the longest word
longest_word = find_longest_word('template.txt')

# print the longest word
print(f"The longest word in the file is: {longest_word} and the number of characters are {len(longest_word)}")
```
